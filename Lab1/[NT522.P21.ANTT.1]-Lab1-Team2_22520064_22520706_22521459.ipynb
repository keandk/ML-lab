# Bài tập về nhà 3: Sử dụng Linear Regression

### Scikit-learning

```python
import numpy as np
from sklearn.linear_model import LinearRegression
# X is a matrix that represents the training dataset
# y is a vector of weights, to be associated with input dataset
X = np.array([[3], [5], [7], [9], [11]]).reshape(-1, 1)
y = [8.0, 9.1, 10.3, 11.4, 12.6]
lreg_model = LinearRegression()
lreg_model.fit(X, y)
# New data (unseen before)
new_data = np.array([[13]])
print('Model Prediction for new data: $%.2f' % lreg_model.predict(new_data)[0])
```

## Tìm hiểu thực hiện lại ví dụ dùng mô hình Linear Regression trong thư viện scikit-learning bằng các thư viện sau:

### TensorFlow

```python
import numpy as np
import tensorflow as tf

X = np.array([[3], [5], [7], [9], [11]], dtype=np.float32)
y = np.array([8.0, 9.1, 10.3, 11.4, 12.6], dtype=np.float32)
# Weight and bias
W = tf.Variable(0.0, dtype=tf.float32)
b = tf.Variable(0.0, dtype=tf.float32)

optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)

for epoch in range(1000):
    with tf.GradientTape() as tape:
        y_pred = W * X + b
        loss = tf.reduce_mean(tf.square(y - y_pred))
    gradients = tape.gradient(loss, [W, b])
    optimizer.apply_gradients(zip(gradients, [W, b]))

# New data (unseen before)
new_data = np.array([[13]], dtype=np.float32)
prediction = (W * new_data + b).numpy()[0]

print(f"TensorFlow Model Prediction: ${prediction.item():.2f}")
```

### Keras

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras

# Data
X = np.array([[3], [5], [7], [9], [11]], dtype=np.float32)
y = np.array([8.0, 9.1, 10.3, 11.4, 12.6], dtype=np.float32)

model = keras.Sequential([
    keras.layers.Dense(1, input_shape=(1,), activation=None)  # Hồi quy tuyến tính
])
model.compile(optimizer='sgd', loss='mse')

# Train
model.fit(X, y, epochs=1000, verbose=0)

# Test
new_data = np.array([[13]], dtype=np.float32)
prediction = model.predict(new_data)[0][0]
print(f"Keras Model Prediction: ${prediction:.2f}")
```

### PyTorch

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Data
X = torch.tensor([[3], [5], [7], [9], [11]], dtype=torch.float32)
y = torch.tensor([[8.0], [9.1], [10.3], [11.4], [12.6]], dtype=torch.float32)

model = nn.Linear(1, 1)
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Train
for epoch in range(1000):
    optimizer.zero_grad()
    y_pred = model(X)
    loss = criterion(y_pred, y)
    loss.backward()
    optimizer.step()

# Test
new_data = torch.tensor([[13]], dtype=torch.float32)
prediction = model(new_data).item()
print(f"PyTorch Model Prediction: ${prediction:.2f}")
```

## Cảm	nghĩ	về	việc	dùng	4	thư	viện	này

1.   Scikit-Learn:
Dễ sử dụng nhất vì có sẵn LinearRegression(), chỉ cần gọi .fit() là xong.
Phù hợp cho các bài toán machine learning đơn giản, không cần tối ưu hóa sâu.
Tuy nhiên, không linh hoạt khi muốn tùy chỉnh quá trình huấn luyện hoặc kiểm soát chi tiết thuật toán tối ưu.
2.   TensorFlow:
Mạnh mẽ cho các mô hình deep learning, nhưng hơi cồng kềnh với bài toán hồi quy tuyến tính đơn giản.
Cần tự định nghĩa trọng số, bias, và vòng lặp huấn luyện.
Kiểm soát được quá trình học, nhưng phải hiểu về GradientTape() và apply_gradients(), khiến việc triển khai phức tạp hơn.
3.   Keras:
Giao diện thân thiện hơn vì có thể dùng API cấp cao như Sequential và Dense.

Không tối ưu bằng Scikit-Learn cho bài toán nhỏ, nhưng dễ mở rộng sang mô hình lớn hơn.
4.   PyTorch:

Nhưng đối với bài toán đơn giản như hồi quy tuyến tính, cách triển khai cũng phức tạp hơn Scikit-Learn.

Tổng kết:
Nếu chỉ cần hồi quy tuyến tính nhanh chóng, Scikit-Learn là lựa chọn tốt nhất. Nếu muốn kiểm soát chi tiết hơn quá trình học, có thể dùng PyTorch hoặc TensorFlow. Còn Keras là lựa chọn trung gian, dễ dùng hơn TensorFlow nhưng vẫn đủ linh hoạt cho các mô hình mở rộng.

# Bài tập 4: Code phát hiện Spam bằng SVMs và Linear Regression

## Bằng SVMs

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
# Đọc file dataset, biến thành các giá trị số
df = pd.read_csv('/content/sms_spam_svm.csv')
y = df.iloc[:, 0].values
y = np.where(y == 'spam', -1, 1)
X = df.iloc[:, [1, 2]].values
# chia dữ liệu thành tập train và tập test
# test_size là 30%
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.3, random_state=0)
svm = SVC(kernel='linear', C=1.0, random_state=0)
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)
print('Misclassified samples: %d' % (y_test != y_pred).sum())
print('Accuracy: %.3f' % accuracy_score(y_test, y_pred))
```

## Bằng Linear Regression

```python
from sklearn.linear_model import LinearRegression
# Đọc file dataset, biến thành các giá trị số
df = pd.read_csv('/content/sms_spam_svm.csv')
y = df.iloc[:, 0].values
y = np.where(y == 'spam', -1, 1)
X = df.iloc[:, [1, 2]].values
# chia dữ liệu thành tập train và tập test
# test_size là 30%
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.3, random_state=0)
# Khởi tạo mô hình hồi quy tuyến tính
lr = LinearRegression()
# Huấn luyện mô hình
lr.fit(X_train, y_train)
# Dự đoán nhãn
y_pred = lr.predict(X_test)
# Làm tròn giá trị dự đoán về -1 hoặc 1
y_pred = [1 if pred >= 0 else -1 for pred in y_pred]
# Đánh giá kết quả
print('Misclassified samples: %d' % (y_test != y_pred).sum())
print('Accuracy: %.3f' % accuracy_score(y_test, y_pred))
```

# Bài tập 5: Phát	hiện	Phishing	với	ML - Logistic regession

```python
import pandas as pd
import numpy as np
from sklearn import *
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
phishing_dataset = np.genfromtxt('/content/drive/MyDrive/Labs/Lab1/phishing_dataset.csv', delimiter=',',dtype=np.int32)
samples = phishing_dataset[:,:-1]
targets = phishing_dataset[:, -1]

from sklearn.model_selection import train_test_split

training_samples, testing_samples, training_targets, testing_targets = train_test_split(samples, targets, test_size=0.2, random_state=0)

log_classifier = LogisticRegression()
log_classifier.fit(training_samples, training_targets)
predictions = log_classifier.predict(testing_samples)

accuracy = 100.0 * accuracy_score(testing_targets, predictions)
print ("Logistic Regression accuracy: " + str(accuracy))
```

## Chức năng của phương thức genfromtxt() trong thư viện numpy

Tải dữ liệu từ tệp văn bản, xử lý các missing values (là dữ liệu bị khuyết, trống hoặc không tồn tại trong một tập dữ liệu) theo chỉ định.
Một số parameters phổ biến như:
*   **fname:** file, str, pathlib.Path, list of str, generator \
Tệp, tên tệp, danh sách,...Nếu phần mở rộng tên tệp là .gzhoặc .bz2, thì tệp sẽ được giải nén trước.
*   **dtype:** dtype, optional \
Kiểu dữ liệu của mảng kết quả. Nếu Không có, các dtype sẽ được xác định theo nội dung của từng cột, riêng lẻ.
*   **delimiter:** str, int, or sequence, optional \
Chuỗi được sử dụng để phân tách các giá trị. Theo mặc định, bất kỳ khoảng trắng liên tiếp nào cũng hoạt động như dấu phân cách. Một số nguyên hoặc chuỗi số nguyên cũng có thể được cung cấp làm chiều rộng của mỗi trường.



Ví dụ về sử dụng phương thức genfromtxt() trong đoạn mã trên:
*   '/content/drive/MyDrive/Labs/Lab1/phishing_dataset.csv': Đường dẫn đến tệp CSV chứa dữ liệu về phishing

*   delimiter=',': Dữ liệu trong tệp CSV được phân tách bằng dấu phẩy , \
*   dtype=np.int32: Chỉ định kiểu dữ liệu là số nguyên 32-bit (int32)

# Bài tập 6: Sinh	viên hoàn thiện code Decision trees trên và đánh giá kết quả nhận được so với	phương pháp Logistic regression.

```python
from sklearn import tree
phishing_dataset = np.genfromtxt('/content/phishing_dataset.csv', delimiter=',',dtype=np.int32)
samples = phishing_dataset[:,:-1]
targets = phishing_dataset[:, -1]
training_samples, testing_samples, training_targets, testing_targets = train_test_split(samples, targets, test_size=0.2, random_state=0)

tree_classifier = tree.DecisionTreeClassifier()
tree_classifier.fit(training_samples, training_targets)
predictions = tree_classifier.predict(testing_samples)
accuracy = 100.0 * accuracy_score(testing_targets, predictions)
print ("Decision Tree accuracy: " + str(accuracy))
```

## So sánh với kết quả nhận được của Logistic Regression
- Decision Tree có độ chính xác cao hơn Logistic Regression (96.47% vs. 91.72%). Điều này có thể do Decision Tree có khả năng học tốt các mẫu phức tạp hơn.
- Logistic Regression hoạt động tốt khi dữ liệu có sự phân tách tuyến tính, trong khi Decision Tree linh hoạt hơn với dữ liệu phi tuyến.
